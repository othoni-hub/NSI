**Pandas par les exemples**

***Pandas*** permet de fabriquer des variables d'un type qui lui est propre : les ***DataFrames***.

À partir de données construites sur place ou, de fichiers au format .csv ou au format .json, enregistrés sur le disque dur ou téléchargés depuis l'internet, on peut traiter ces données et les analyser.

Les DataFrames sont des tableaux, indexés ou non, étiquetés ou non, de données de types multiples, éventuellement très massifs.

Une colonne d'un DataFrame s'appelle une ***Série***, la plupart de ces séries portent un nom en 1<sup>ère</sup> ligne (ligne n°0) du DataFrame. La "Série" est d'ailleurs en soi un type construit de données de Pandas : unidimensionnel et homogène.

Chaque ligne d'un DataFrame est repérée par un ***Index***, correspondant à chaque instance ("individu") de la base de données. Ces deux termes, ***Série*** et ***Index*** sont très importants !

Pandas dispose de fonctions et de méthodes pour en faire des sous-tableaux, les trier horizontalement ou verticalement, les fusionner de diverses façons, opérer sur les données, etc, et surtout, en effectuer des traitements statistiques. Il dispose aussi de quelques fonctionnalités graphiques de base, mais qui parfois peuvent suffire.

Nous allons en explorer quelques unes des principales.

Quant à ***Matplotlib***, il s'agit d'une bibliothèque permettant de réaliser de très nombreuses espèces de graphiques autour de ces données, car, à l'instar du *Data Scientist*, en matière d'analyse de données, la première chose à effectuer est ce que l'on nomme le pré-traitement : nettoyage et visualisation des données en 1,2, ou 3 dimensions. Au-delà, on réduit éventuellement le nombre de dimensions (de variables présentes) et la taille de ces données, avec des outils tels les classifieurs ou l'Analyse en Composantes Principales (ACP), afin de pouvoir s'en faire la meilleure idée possible.

Créons une base de données de 11 pays avec leurs émissions de CO2 et leur population, ainsi qu'une colonne pour le continent auquel ils appartiennent :

```
import numpy as np # juste pour quelques laisser quelques données vides...
import pandas as pd

df_CO2 = pd.DataFrame({
    'Country':['USA','China','India','Russia','Switzerland','Japan','Sweden','Singapore','South Korea','UK','Australia'],
    'Co2_emission':[5107.393,10877.218,2454.774,1764.866,39.738,1320.776,np.NaN,55.018,673.324,379.150,764.866],
    'Population_million':[329,1433,1366,145,8.5,126,10,5.8,51,67,np.NaN],
    'Continent': ['NA','Asia','Asia','EU','EU','Asia','EU','Asia','Asia','EU','AU']
                 })

df_CO2
```

### **a/ Trier par valeur (= sur une variable)**

Ici, nous trions la base de données par valeur de colonne, en l'occurrence  *population_million*

## **4.Plusieurs façons de fusionner des données de DataFrames**

source : https://towardsdatascience.com/combining-dataframes-using-pandas-b9e2e83b9869 

Fabriquons de toute pièce deux petits DataFrames :

```
df1 = pd.DataFrame({'column_a' : [1,2,3,4], 
                    'column_b' : ['a','b','c','d'],
                    'column_c' : [True, True, False, True]})

df2 = pd.DataFrame({'column_a' : [1,2,9,10], 
                    'column_b' : ['a','k','l','m'],
                    'column_c' : [False, False, False, True]})
df1 
```

```
df2
```

### **a/ Concaténation**

Une première façon de combiner deux DataFrames est la fonction ***concat()***.
Sans autre précision, elle met simplement bout à bout les deux tableaux, verticalement, en ajoutant des lignes, quitte à obtenir des doublons sur les index.

```
df = pd.concat([df1, df2])
df
```

Pour réindexer les lignes, on rajoute l'attribut *ignore_index = True* :

```
df = pd.concat([df1, df2], ignore_index = True)
df
```

En précisant *axis = 1*, on va accoler les deux tableaux en largeur, en ajoutant des colonnes, quitte à créer des doublons dans les entêtes de colonnes.
Si deux lignes portent le même index, elles sont fusionnées, quitte à compléter les cases vides par des *NaN*

```
df = pd.concat([df1, df2], axis = 1)
df
```

### **b/ Jointure**

Ce qui distingue la jointure de la concaténation est l'élimination des doublons.
Si l'option ***inner*** est choisie, seules les lignes avec des indices communs seront jointes, quitte à obtenir des doublons dans les noms de colonnes.


Réindexons le 2<sup>ème</sup> DataFrame pour que l'illustration soit plus parlante...

```
df2 = pd.DataFrame({'column_a' : [1,2,9,10], 
                    'column_b' : ['a','k','l','m'],
                    'column_c' : [False, False, False, True]}, index = [2,3,4,5])

df2
```

```
df = pd.concat([df1, df2], axis = 1, join = 'inner')
df
```

Si le paramètre *join* est placé à ***outer*** (sa valeur par défaut), tous les index figurent dans le DataFrame final, quitte à remplir les case vides par des *NaN* 

```
df = pd.concat([df1, df2], axis = 1, join = 'outer')
df
```

Pandas permet également de rajouter des étiquettes pour indiquer de quel DataFrame proviennent chaque part :

```
df = pd.concat([df1, df2], keys = ['df1','df2'])
df
```

Ceci permet d'accéder plus facilement aux diverse parties du DataFrame final :

```
df.loc['df2']
```

**À retenir :** La fonction ***concat()*** fabrique une copie des données !

### **c/ Fusion**

Là où la concaténation se contentait de mettre bout à bout les données, la fusion (***merge()***) va combiner de manière plus flexible les DataFrames en se basant sur les valeurs dans les colonnes partagées.
Le paramètre **on** indique sur quelle(s) colonne(s) va se situer la fusion.

```
df = pd.merge(df1, df2, on = "column_a")
df
```

Ici, par défaut, seules les lignes communes, dans les valeurs de la colonne sur laquelle s'opère la fusion, sont conservées.
Cela se règle par un paramètre ***how***, réglé par défaut sur ***inner***.
Pour conserver toutes les lignes, on règle ce paramètre à ***outer***.
Les cases vides sont alors fixées à NaN.

```
df = pd.merge(df1, df2, on = "column_a", how = 'outer')
df
```

### **d/ Pivot (tableau croisé dynamique)**

Cette méthode consiste à créer une seconde table en utilisant les valeurs d’une colonne comme nom de colonnes.

### **a/ Chargement et fusion des données**

Les données sont issues de deux *datasets*, concernant les vins rouges et les vins blancs, que nous allons fusionner

```
white_wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv', sep=';')
red_wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv', sep=';')

# store wine type as an attribute
red_wine['wine_type'] = 'red'   
white_wine['wine_type'] = 'white'

# bucket wine quality scores into qualitative quality labels
red_wine['quality_label'] = red_wine['quality'].apply(lambda value: 'low' 
                                                          if value <= 5 else 'medium' 
                                                              if value <= 7 else 'high')
red_wine['quality_label'] = pd.Categorical(red_wine['quality_label'], 
                                           categories=['low', 'medium', 'high'])
white_wine['quality_label'] = white_wine['quality'].apply(lambda value: 'low' 
                                                              if value <= 5 else 'medium' 
                                                                  if value <= 7 else 'high')
white_wine['quality_label'] = pd.Categorical(white_wine['quality_label'], 
                                             categories=['low', 'medium', 'high'])

# merge red and white wine datasets
wines = pd.concat([red_wine, white_wine])

# re-shuffle records just to randomize data points
wines = wines.sample(frac=1, random_state=42).reset_index(drop=True)

wines.head()
```

```
print('Le jeu de données comporte ',len(wines),' vins.')
```

**Comprendre le vin**

Le vin est une boisson alcoolisée fabriquée à partir de raisins qui est fermentée sans ajout de sucres, d'acides, d'enzymes, d'eau ou d'autres nutriments.

Le vin rouge est fabriqué à partir de raisins rouges et noirs foncés. La couleur varie généralement de diverses nuances de rouge, de brun et de violet. Il est produit à partir de raisins entiers, y compris la peau, ce qui ajoute à la couleur et à la saveur des vins rouges et leur donne un goût riche.

Le vin blanc est produit à partir de raisins blancs sans peau ni pépins. La couleur est généralement jaune paille, jaune vert ou jaune or. La plupart des vins blancs ont une saveur légère et fruitée par rapport aux vins rouges plus riches.

**Comprendre les attributs et les propriétés du vin**
* une acidité fixe (***fixed acidity***) en g/dm<sup>3</sup> : Les acides sont l'une des propriétés fondamentales du vin et contribuent grandement à son goût. Une réduction significative des acides peut conduire à un goût plat du vin. Les acides fixes comprennent les acides tartrique, malique, citrique et succinique que l'on trouve dans le raisin (sauf le succinique).


* l'acidité volatile (***volatile acidity***) en g/dm<sup>3</sup> : Ces acides doivent être distillés du vin avant de terminer le processus de production. Ils sont principalement constitués d'acide acétique, bien que d'autres acides comme les acides lactique, formique et butyrique puissent également être présents. L'excès d'acides volatils est indésirable et entraîne une saveur désagréable. 

* l'acide citrique (***citric acid***) en g/dm<sup>3</sup> : C'est l'un des acides fixes qui donnent au vin sa fraîcheur. Il est généralement consommé en grande partie pendant le processus de fermentation et parfois ajouté séparément pour donner plus de fraîcheur au vin.

* sucre résiduel (***residual sugar***) en g/dm<sup>3</sup> : Il s'agit généralement du sucre naturel du raisin qui reste après l'arrêt du processus de fermentation. Il est généralement exprimé en $\frac{g}{dm^3}$ dans l'ensemble de données.

* chlorures (***chlorides***) en g/dm<sup>3</sup> : C'est généralement un élément majeur de la salinité du vin. 

* dioxyde de soufre libre (***free sulfur dioxide***) en mg/dm<sup>3</sup> : C'est la partie du dioxyde de soufre qui, lorsqu'elle est ajoutée à un vin, est dite libre après que la partie restante se soit liée. Les viticulteurs s'efforceront toujours d'obtenir la plus grande proportion de soufre libre pour se lier. On les appelle aussi sulfites et une trop grande quantité est indésirable et donne une odeur piquante (et mal à la tête...)

* le dioxyde de soufre total (***total sulfur dioxide***) en mg/dm<sup>3</sup> : C'est la somme totale du dioxyde de soufre lié et du dioxyde de soufre libre ($SO_2$). Il est principalement ajouté pour tuer les bactéries nocives et préserver la qualité et la fraîcheur. Il existe généralement des limites légales pour les niveaux de soufre dans les vins et un excès peut même tuer de bonnes levures et dégager une odeur indésirable.

* densité (***density***) en g/cm<sup>3</sup>: Elle peut être représentée comme une comparaison du poids d'un volume spécifique de vin à un volume équivalent d'eau. Elle est généralement utilisée comme mesure de la conversion du sucre en alcool.

* le **pH** : Il s'agit d'une échelle numérique permettant de préciser l'acidité ou la basicité du vin. L'acidité fixe est celle qui contribue le plus au pH des vins. Vous le savez peut-être, les solutions dont le pH est inférieur à 7 sont acides, tandis que les solutions dont le pH est supérieur à 7 sont basiques. Avec un pH de 7, l'eau pure est neutre. La plupart des vins ont un pH compris entre 2,9 et 3,9 et sont donc acides.


* sulfates (***sulphates***) en g/dm<sup>3</sup>: Ce sont des sels minéraux contenant du soufre. Les sulfates sont au vin ce que le gluten est à la nourriture. Ils font partie intégrante de la vinification dans le monde entier et sont considérés comme essentiels. Ils sont liés au processus de fermentation et affectent l'arôme et la saveur du vin. 

* l'alcool (***alcohol***) en % du volume : Le vin est une boisson alcoolisée. L'alcool est formé par la conversion du sucre par la levure pendant le processus de fermentation. Le pourcentage d'alcool peut varier d'un vin à l'autre. Il n'est donc pas surprenant que cet attribut fasse partie de cet ensemble de données. 

* qualité (***quality***) : Les experts en vin ont noté la qualité du vin entre 0 (très mauvais) et 10 (excellent). La note finale de qualité est la médiane d'au moins trois évaluations faites par les mêmes experts en vin.

* type_de_vin (***wine_type***) : Comme nous avions à l'origine deux ensembles de données pour le vin rouge et le vin blanc, nous avons introduit cet attribut dans l'ensemble de données fusionné final qui indique le type de vin pour chaque point de données. Un vin peut être soit un vin "rouge", soit un vin "blanc". 

* label_qualité (***quality_label***) : Il s'agit d'un attribut dérivé de l'attribut de qualité. Nous classons ou regroupons les notes de qualité du vin en trois catégories qualitatives : faible, moyenne et élevée. Les vins ayant une note de qualité de 3, 4 et 5 sont de faible qualité, les notes de 6 et 7 sont de qualité moyenne et les notes de 8 et 9 sont des vins de haute qualité.
